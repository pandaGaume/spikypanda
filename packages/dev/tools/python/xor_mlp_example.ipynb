
# XOR Neural Network Example (2-2-1 MLP)

This notebook implements a Multi-Layer Perceptron (MLP) to learn the XOR function using PyTorch.

## XOR Dataset


```python

import torch
import torch.nn as nn
import torch.optim as optim

#torch.manual_seed(42)  #force reproductible random values..

# XOR dataset
xor_inputs = torch.tensor([[0., 0.], [0., 1.], [1., 0.], [1., 1.]])
xor_targets = torch.tensor([[0.], [1.], [1.], [0.]])

# Define the MLP model
class XORNet(nn.Module):
    def __init__(self):
        super(XORNet, self).__init__()
        self.hidden = nn.Linear(2, 2)
        self.output = nn.Linear(2, 1)
        self.activation_hidden = nn.Tanh()
        self.activation_output = nn.Sigmoid()
        # Glorot/Xavier initialization
        #nn.init.xavier_uniform_(self.hidden.weight)
        #nn.init.zeros_(self.hidden.bias)
        #nn.init.xavier_uniform_(self.output.weight)
        #nn.init.zeros_(self.output.bias)

    def forward(self, x):
        x = self.activation_hidden(self.hidden(x))
        x = self.activation_output(self.output(x))
        return x

# Initialize model, loss, and optimizer
model = XORNet()
criterion = nn.BCELoss()
optimizer = optim.SGD(model.parameters(), lr=0.5)

print("Initials Hidden weights:", model.hidden.weight.data)
print("Initials Output weights:", model.output.weight.data)

# Training loop
epochs = 10000
for epoch in range(epochs):
    optimizer.zero_grad()
    outputs = model(xor_inputs)
    loss = criterion(outputs, xor_targets)
    loss.backward()
    optimizer.step()
    if epoch % 1000 == 0:
        print(f"Epoch {epoch} — Loss: {loss.item()}")

# Final predictions
print("\nFinal predictions:")
with torch.no_grad():
    predictions = model(xor_inputs)
    for i, pred in enumerate(predictions):
        print(f"Input: {xor_inputs[i].tolist()}, Predicted: {pred.item():.4f}, Expected: {xor_targets[i].item()}")

# Extract and print final weights and biases
print("\nFinal weights and biases:")
print("Hidden weights:", model.hidden.weight.data)
print("Hidden biases:", model.hidden.bias.data)
print("Output weights:", model.output.weight.data)
print("Output biases:", model.output.bias.data)
```
